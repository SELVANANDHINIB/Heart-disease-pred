```python
import pandas as pd
import sklearn
```

Data Collection and reading the data using pandas library

```python
dataset = pd.read_csv('heart.csv')
```

```python
dataset
```

```python
X=dataset.drop(columns='target',axis=1)
```

```python
X
```

```python
Y=dataset[['target']]
```

```python
Y
```

Splitting of training and testing data using module train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.75,random_state=None)
```

```python
x_train
```

```python
x_test
```

```python


y_train
```

```python
y_test
```

Random Forest Algorithm:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ensemble=RandomForestClassifier()
```

```python
ensemble.fit(x_train,y_train)
```

```python
pre=ensemble.predict(x_test)
```

```python
pre
```

```python
from sklearn.metrics import accuracy_score

# Assuming pred and y_test are already defined
print("Accuracy:", accuracy_score(pre, y_test))
```
